```python
import numpy as np
import matplotlib.pyplot as plt

def plot_crs_vrs(mel_df, input_col, output_col, ax, title=""):
    """
    Vykreslí do Axes 'ax' graf so vstupom 'input_col' (X-ová os)
    a výstupom 'output_col' (Y-ová os), doplnený o:
      - CRS priamku (oranžová)
      - VRS (čierna prerušovaná) podľa is_efficient
      - body DMU rozdelené na efektívne / neefektívne
    """
    # 1) Rozdelenie na efektívne/neefektívne DMU
    df_eff = mel_df[mel_df['is_efficient'] == 1].copy()
    df_ineff = mel_df[mel_df['is_efficient'] == 0].copy()

    # 2) CRS priamka:
    #    - nájdeme DMU s najlepším pomerom (output / input)
    best_ratio = 0
    best_x, best_y = None, None

    for _, row in mel_df.iterrows():
        xi = row[input_col]
        yi = row[output_col]
        if xi > 0:
            ratio = yi / xi
            if ratio > best_ratio:
                best_ratio = ratio
                best_x = xi
                best_y = yi

    # 3) VRS krivka:
    #    - zoberieme DMU s is_efficient==1, zoradíme podľa vstupu (input_col)
    df_eff_sorted = df_eff.sort_values(by=input_col)

    # 4) Vykreslenie
    # 4a) Neefektívne DMU
    ax.plot(
        df_ineff[input_col],
        df_ineff[output_col],
        "o",
        label="not-efficient DMU"
    )
    # 4b) Efektívne DMU + prerušovaná čiara
    ax.plot(
        df_eff_sorted[input_col],
        df_eff_sorted[output_col],
        "o--",
        color="black",
        label="VRS surface"
    )
    # 4c) Oranžová priamka pre CRS
    #    od (0,0) k (best_x, best_y)
    ax.plot(
        [0, best_x],
        [0, best_y],
        color="orange",
        label="CRS surface"
    )

    ax.set_xlabel(input_col)
    ax.set_ylabel(output_col)
    ax.set_title(title)
    ax.grid(True)


def plot_all_four(mel_df):
    """
    Vytvorí 4 grafy (2×2) pre:
      - O1 vs I1
      - O2 vs I1
      - O3 vs I1
      - (O1 + O2 + O3) vs I1
    """
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    # Graf 1: O1
    plot_crs_vrs(mel_df, "I1", "O1", axs[0,0], title="I1 vs O1")

    # Graf 2: O2
    plot_crs_vrs(mel_df, "I1", "O2", axs[0,1], title="I1 vs O2")

    # Graf 3: O3
    plot_crs_vrs(mel_df, "I1", "O3", axs[1,0], title="I1 vs O3")

    # Graf 4: súčet O1 + O2 + O3
    mel_df['O_sum'] = mel_df['O1'] + mel_df['O2'] + mel_df['O3']
    plot_crs_vrs(mel_df, "I1", "O_sum", axs[1,1], title="I1 vs (O1+O2+O3)")

    # Legenda pre všetky subpl
    handles, labels = axs[0,0].get_legend_handles_labels()
    fig.legend(handles, labels, loc="upper center", ncol=3)

    plt.tight_layout()
    plt.show()


# Hlavná časť: zavolaj funkciu plot_all_four(mel_df)
# Predpokladáme, že 'mel_df' obsahuje stĺpce: I1, O1, O2, O3, is_efficient
plot_all_four(mel_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_dea_frontier(df, output_col=None, output_option="single", agg_outputs=None, title=None):
    """
    Plots a DEA output-oriented frontier for a stable input (I1) against a selected output.
    
    Parameters:
      df: DataFrame containing at least the following columns:
          - I1: Input (stable)
          - O1, O2, O3: Output columns (or others if specified via agg_outputs)
          - is_efficient: Indicator (1/True for efficient DMU, 0/False otherwise)
          - (and possibly other columns)
      output_col: Name of the output column to plot (if output_option is "single").
                  This parameter is ignored if output_option is "mean" or "sum".
      output_option: A string that determines which output to use:
                     - "single": use the column specified by output_col.
                     - "mean": compute the mean of the specified output columns.
                     - "sum": compute the sum of the specified output columns.
      agg_outputs: Optional list of output columns to aggregate when output_option is "mean" or "sum".
                   If not provided, defaults to ["O1", "O2", "O3"].
      title: Optional title for the plot. If not provided, a default title is used.
    """
    if agg_outputs is None:
        agg_outputs = ["O1", "O2", "O3"]
    
    # Decide on the output column based on output_option
    if output_option == "mean":
        # Compute the mean of the specified output columns and store it in a new column "O_mean"
        df = df.copy()
        df["O_mean"] = df[agg_outputs].mean(axis=1)
        used_output = "O_mean"
    elif output_option == "sum":
        # Compute the sum of the specified output columns and store it in a new column "O_sum"
        df = df.copy()
        df["O_sum"] = df[agg_outputs].sum(axis=1)
        used_output = "O_sum"
    elif output_option == "single":
        if output_col is None:
            raise ValueError("When using output_option 'single', you must specify an output_col.")
        used_output = output_col
    else:
        raise ValueError("output_option must be one of: 'single', 'mean', or 'sum'.")
    
    # Compute summary statistics for I1 (input) and the selected output
    minx = df['I1'].min()
    maxx = df['I1'].max()
    miny = df[used_output].min()
    maxy = df[used_output].max()
    
    # Separate efficient and non-efficient DMU
    df_eff = df[df['is_efficient'] == 1].copy()
    df_ineff = df[df['is_efficient'] == 0].copy()
    
    # Construct CRS Frontier (output-oriented):
    # For output-oriented model, the efficient frontier is represented by the best ratio (output / I1).
    best_ratio = 0
    best_x, best_y = 0, 0
    for _, row in df.iterrows():
        xi = row['I1']
        yi = row[used_output]
        if xi > 0:
            ratio = yi / xi
            if ratio > best_ratio:
                best_ratio = ratio
                best_x = xi
                best_y = yi
    crs_x = [0, best_x]
    crs_y = [0, best_y]
    
    # Construct VRS Frontier:
    # Sort the efficient DMU by I1 in ascending order and extract the corresponding output values.
    df_eff_sorted = df_eff.sort_values(by='I1')
    vrs_x = list(df_eff_sorted['I1'])
    vrs_y = list(df_eff_sorted[used_output])
    
    # Create a polygon for filling (optional): add lower bound points to close the area.
    poly_x = [minx] + vrs_x + [maxx, minx]
    poly_y = [miny] + vrs_y + [miny, miny]
    
    # Create the plot
    plt.figure(figsize=(8,6))
    
    # Fill the area under the efficient (VRS) frontier with a light transparent orange
    plt.fill(poly_x, poly_y, color="orange", alpha=0.15, label="Efficient area (CRS)")
    
    # Plot the VRS frontier as a dashed black line
    plt.plot(vrs_x, vrs_y, "o--", color="black", label="VRS frontier")
    
    # Plot the CRS frontier as a solid orange line
    plt.plot(crs_x, crs_y, color="orange", label="CRS frontier")
    
    # Scatter plots of DMU
    # Efficient DMU (blue markers)
    plt.scatter(df_eff['I1'], df_eff[used_output], c="blue", s=80, marker="o", label="Efficient DMU")
    # Non-efficient DMU (gray markers)
    plt.scatter(df_ineff['I1'], df_ineff[used_output], c="gray", s=40, marker="o", label="Not-efficient DMU")
    
    # Add reference lines (example values, adjust as needed)
    plt.plot([6, 6], [2, 0], color="C0", label="Reference line 1")
    plt.plot([2, 0], [5, 5], color="C0", label="Reference line 2")
    plt.plot([0, 3.6], [0, 4.2], color="black", linestyle="--", label="Reference line 3")
    
    plt.xlabel("Input (I1)")
    plt.ylabel(f"Output ({used_output})")
    if title is None:
        title = f"DEA Output-Oriented Frontier Visualization: I1 vs {used_output}"
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xlim(0, maxx * 1.1)
    plt.ylim(0, maxy * 1.1)
    plt.show()

# Example calls:
# Plot for I1 vs O1 (single output)
plot_dea_frontier(mel_df, output_col="O1", output_option="single", 
                  title="I1 vs O1 Frontier")

# Plot for I1 vs mean of selected outputs (e.g., O1, O2)
plot_dea_frontier(mel_df, output_option="sum", agg_outputs=["O1", "O2", "O3"],
                  title="I1 vs Mean of O1, O2 and O3 Frontier")

# Plot for I1 vs sum of selected outputs (e.g., O1, O2, O3)
plot_dea_frontier(mel_df, output_option="sum", agg_outputs=["O1", "O2"],
                  title="I1 vs Sum of O1, O2 Frontier")
```

```python
import plotly.graph_objects as go
import plotly.io as pio

def plot_dmu_3d(df, input_col, output_col1, output_col2, title, export_filename=None):
    """
    Creates a 3D scatter plot for DMU with one input and two outputs.
    Each point is labeled with the project name from the 'Projekt' column.
    
    Parameters:
      df: DataFrame containing columns for input, outputs, 'Projekt', and 'is_efficient'
      input_col: Name of the input column (e.g., "I1")
      output_col1: Name of the first output column (e.g., "O1")
      output_col2: Name of the second output column (e.g., "O3")
      title: Title for the plot
      export_filename: Optional filename (string) to export the figure as an HTML file.
    """
    # Separate efficient and non-efficient DMU
    df_eff = df[df['is_efficient'] == 1].copy()
    df_ineff = df[df['is_efficient'] == 0].copy()

    fig = go.Figure()

    # Efficient DMU: blue markers with project labels
    fig.add_trace(
        go.Scatter3d(
            x=df_eff[input_col],
            y=df_eff[output_col1],
            z=df_eff[output_col2],
            mode='markers+text',
            marker=dict(size=7, color='blue'),
            text=df_eff['Projekt'],
            textposition="top center",
            name='Efficient DMU'
        )
    )

    # Not-efficient DMU: gray markers with project labels
    fig.add_trace(
        go.Scatter3d(
            x=df_ineff[input_col],
            y=df_ineff[output_col1],
            z=df_ineff[output_col2],
            mode='markers+text',
            marker=dict(size=5, color='gray'),
            text=df_ineff['Projekt'],
            textposition="top center",
            name='Not-efficient DMU'
        )
    )

    fig.update_layout(
        title=title,
        scene=dict(
            xaxis_title=f'{input_col} (Input)',
            yaxis_title=f'{output_col1} (Output)',
            zaxis_title=f'{output_col2} (Output)'
        ),
        legend=dict(x=0, y=1.0)
    )
    
    # Show the figure
    fig.show()
    
    # If export_filename is provided, export the figure as HTML
    if export_filename:
        pio.write_html(fig, file=export_filename, auto_open=True)

# Example calls:
plot_dmu_3d(mel_df, 'I1', 'O1', 'O2', "DEA 3D Visualization: I1 vs O1 and O2", export_filename="dea_3d_I1_O1_O2.html")
plot_dmu_3d(mel_df, 'I1', 'O1', 'O3', "DEA 3D Visualization: I1 vs O1 and O3", export_filename="dea_3d_I1_O1_O3.html")
plot_dmu_3d(mel_df, 'I1', 'O2', 'O3', "DEA 3D Visualization: I1 vs O2 and O3", export_filename="dea_3d_I1_O2_O3.html")
```
